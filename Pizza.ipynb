```python
class pizza(baska_bir_class):
  sozluk_pizza = {"İsim":[],"Değer":[],"Sos":[],"Malzeme":[]}

  def __init__(self,isim,deger,sos,malzeme):
    self.isim = isim
    self.deger = deger
    self.sos = sos
    self.malzeme = malzeme

    if isim not in self.sozluk_pizza["İsim"]:
      self.sozluk_pizza["İsim"].append(isim)
      self.sozluk_pizza["Değer"].append(deger)
      self.sozluk_pizza["Sos"].append(sos)
      self.sozluk_pizza["Malzeme"].append(malzeme)

    else:
      print("Bu pizza kayıtlı, bilgiler değiştirildi sadece")
      index_deger = self.sozluk_pizza["İsim"].index(isim)
      self.sozluk_pizza["Değer"][index_deger] = deger
      self.sozluk_pizza["Sos"][index_deger] = sos
      self.sozluk_pizza["Malzeme"][index_deger] = malzeme

      print("İsim \t"+self.sozluk_pizza["İsim"][index_deger])
      print("Değer \t"+str(self.sozluk_pizza["Değer"][index_deger]))
      print("Sos"+self.sozluk_pizza["Sos"][index_deger])
      print(self.sozluk_pizza["Malzeme"][index_deger])
  
  def listeleme():
    print("İsim\t\t Değer\t Sos\t\t Malzeme\n")
    for i in range(len(pizza.sozluk_pizza["İsim"])):
      print("{}\t {}\t {}\t {}".format(pizza.sozluk_pizza["İsim"][i],pizza.sozluk_pizza["Değer"][i],pizza.sozluk_pizza["Sos"][i],pizza.sozluk_pizza["Malzeme"][i]))

  def csv_olustur():
    column_isim = ["İsim","Değer","Sos","Malzeme"]
    with open('pizza.csv', 'a') as csv_file:
        yazdirma_islem = csv.writer(csv_file)

        yazdirma_islem.writerow(column_isim)

    for i in range(len(pizza.sozluk_pizza["İsim"])):
      with open('pizza.csv', 'a') as csv_file:
          yazdirma_islem = csv.writer(csv_file)
          yazdirma_islem.writerow([pizza.sozluk_pizza["İsim"][i],pizza.sozluk_pizza["Değer"][i],pizza.sozluk_pizza["Sos"][i],pizza.sozluk_pizza["Malzeme"][i]])
```
